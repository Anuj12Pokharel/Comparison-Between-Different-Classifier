````python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

# Define the objective function to optimize
def objective_function(X, X_train, Y_train, X_test, Y_test):
    rfc = RandomForestClassifier(n_estimators=int(X[0]), max_depth=int(X[1])) # Convert the hyperparameters to integers
    rfc.fit(X_train, Y_train)
    pred = rfc.predict(X_test)
    return -accuracy_score(Y_test, pred)  # Negative of accuracy score since PSO minimizes

# Define PSO algorithm
def particle_swarm_optimization(obj_function, bounds, X_train, Y_train, X_test, Y_test, n_particles=10, n_iterations=100, w=0.5, c1=2, c2=2):
    n_variables = len(bounds)
    # Initialize particles' positions and velocities
    particles_position = np.random.uniform(bounds[:, 0], bounds[:, 1], (n_particles, n_variables))
    particles_velocity = np.zeros((n_particles, n_variables))
    # Initialize best particle position and global best position
    best_particle_position = particles_position.copy()
    global_best_position = particles_position[np.argmin([obj_function(p, X_train, Y_train, X_test, Y_test) for p in particles_position])]
    # Iterate through iterations
    for _ in range(n_iterations):
        # Update velocities and positions
        particles_velocity = w*particles_velocity + c1*np.random.rand()*(best_particle_position - particles_position) + c2*np.random.rand()*(global_best_position - particles_position)
        particles_position += particles_velocity
        # Clip particle positions to be within bounds
        particles_position = np.clip(particles_position, bounds[:, 0], bounds[:, 1])
        # Update best particle position and global best position
        for i in range(n_particles):
            if obj_function(particles_position[i], X_train, Y_train, X_test, Y_test) < obj_function(best_particle_position[i], X_train, Y_train, X_test, Y_test):
                best_particle_position[i] = particles_position[i].copy()
        new_global_best_position = particles_position[np.argmin([obj_function(p, X_train, Y_train, X_test, Y_test) for p in particles_position])]
        if obj_function(new_global_best_position, X_train, Y_train, X_test, Y_test) < obj_function(global_best_position, X_train, Y_train, X_test, Y_test):
            global_best_position = new_global_best_position.copy()
    return global_best_position

# Define bounds for hyperparameters (e.g., number of trees and max depth)
bounds = np.array([[10, 100], [1, 10]])  # Example bounds for number of trees and max depth

# Perform PSO to optimize hyperparameters
best_hyperparameters = particle_swarm_optimization(objective_function, bounds, X_train, Y_train, X_test, Y_test)

# Train RandomForestClassifier with the best hyperparameters
best_rfc = RandomForestClassifier(n_estimators=int(best_hyperparameters[0]), max_depth=int(best_hyperparameters[1]))
best_rfc.fit(X_train, Y_train)

# Evaluate RandomForestClassifier on test dataset
pred = best_rfc.predict(X_test)
test_accuracy = accuracy_score(Y_test, pred)
print("Best hyperparameters:", best_hyperparameters)
print("Accuracy on test dataset:", test_accuracy)
````
